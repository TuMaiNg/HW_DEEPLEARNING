```python
# Ct lt mng
arr = np.array([1, 2, 3, 4, 5])

sub_arr = arr[1:4]  # Ct t phn t th 1 n 4
print(sub_arr) # 2 3 4 5

sub_arr = arr[:-1]  # 1 2 3 4
print(sub_arr)

sub_arr = arr[-2:]  # 4 5
print(sub_arr)
```

```python
# Chuyn v mng
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
#  chuyn v mng, chng ta c th s dng .T hoc hm numpy.transpose().
transposed_arr = arr.T
print(transposed_arr)
```

```python
# Ni mng
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
# - Hm numpy.sum() cho php tnh tng cc phn t trong mng.
arr = np.array([1, 2, 3, 4, 5])
total = np.sum(arr)
print("Tng ca mng arr:", total)

# III. Hm tnh trung bnh (Mean)
# - Hm numpy.mean() tnh gi tr trung bnh ca cc phn t trong mng.
average = np.mean(arr)
print("Gi tr trung bnh ca mng arr:", average)

# IV. Hm tm gi tr ln nht v nh nht (Max v Min)
# - Hm numpy.max() v numpy.min() dng  tm gi tr ln nht v nh nht trong mng.
max_value = np.max(arr)
min_value = np.min(arr)
print("Gi tr ln nht trong mng arr:", max_value)
print("Gi tr nh nht trong mng arr:", min_value)

# V. Hm tnh  lch chun (Standard Deviation)
# - Hm numpy.std() tnh  lch chun ca mng, o lng mc  phn tn ca d liu.
std_deviation = np.std(arr)
print(" lch chun ca mng arr:", std_deviation)

# VI. Hm tnh phng sai (Variance)
# - Hm numpy.var() tnh phng sai ca mng, o lng mc  bin thin ca d liu.
variance = np.var(arr)
print("Phng sai ca mng arr:", variance)

# VII. Hm tnh tng tch chp (Dot Product)
# - Hm numpy.dot() tnh tng tch chp ca hai mng (vector).
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
dot_product = np.dot(arr1, arr2)
print("Tng tch chp ca arr1 v arr2:", dot_product)
```

# **IV. Gioi thieu ve Scipy**

```python
from scipy import linalg
```

```python
A = np.array([[3,4],[8,9]])
b = np.array([5,7])
```

```python
x = linalg.solve(A,b)
print(f"Ket qua {x}")
```

```python
from scipy import integrate

# Tnh tch phn ca mt hm s
# nh ngha hm f(x) = x^2
def my_function(x):
    return x**2
# Tnh tch phn ca f(x) t 0 n 1
integral = integrate.quad(my_function, 0, 1);
print("Tch phn f(x) t 0 n 1: ", integral)
```

```python
# Tnh gi tr ring v vector ring ca mt ma trn:
A = np.array([[2,1], [3,2]])

# Tnh gi tr ring v vector ring ca ma trn
evals, evecs = linalg.eig(A)

print("Gi tr ring: ", evals)
print("Vector ring: ", evecs)
```
